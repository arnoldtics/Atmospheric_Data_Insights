```python
import pandas as pd
import numpy as	np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
PATH = "../../DataCleaning/"
```

```python
df = pd.read_csv(PATH + "Atmospheric Data Cleaned Instances Per Day.csv")
```

```python
df["CO"] = df["CO"].apply(lambda x: x*1000)
```

## Regression lines with the following variables:
- O3 -> PM10
- O3 -> PM2.5
- O3 -> Temp_Avg
- O3 -> RH_Avg     *\*strong negative correlation*
- O3 -> Rad_Avg
- NO2 -> NO
- NO2 -> WSpeed_Avg
- NO2 -> WDir_Avg
- NO -> WDir_Avg
- NO -> WDir_Avg
- PM10 -> PM2.5     *\*strong positive correlation*
- PM10 -> Temp_Avg
- PM10 -> RH_Avg     *\*strong negative correlation*
- PM2.5 -> Temp_Avg
- PM2.5 -> RH_Avg

## O3 vs PM10

```python
plt.figure(figsize=(14,12))
g = sns.lmplot(x="PM10", y="O3", data=df, markers=".", scatter_kws=dict(color='c', s=10))
g.set_xlabels("PM10\nug/m^3", fontsize=10)
g.set_ylabels("O3\nppb", fontsize=10)
plt.ylim(0,63)
plt.xlim(0,58)
plt.title("Regression plot between O3 and PM10", fontsize=14, pad=20)
plt.show()
```

## O3 vs PM2.5

```python
plt.figure(figsize=(14,12))
g = sns.lmplot(x="PM2.5", y="O3", data=df, markers=".", scatter_kws=dict(color='c', s=10))
g.set_xlabels("PM2.5\nug/m^3", fontsize=10)
g.set_ylabels("O3\nppb", fontsize=10)
plt.ylim(0,63)
plt.xlim(0,58)
plt.title("Regression plot between O3 and PM2.5", fontsize=14, pad=20)
plt.show()
```

## O3 vs Temp_Avg

```python
plt.figure(figsize=(14,12))
g = sns.lmplot(x="Temp_Avg", y="O3", data=df, markers=".", scatter_kws=dict(color='c', s=10))
g.set_xlabels("Temp_Avg\n°C", fontsize=10)
g.set_ylabels("O3\nppb", fontsize=10)
plt.ylim(0,63)
plt.xlim(0,30)
plt.title("Regression plot between O3 and Temp_Avg", fontsize=14, pad=20)
plt.show()
```

## O3 vs RH_Avg

**Strong negative correlation with 99% of confidence**

```python
plt.figure(figsize=(14,12))
g = sns.lmplot(x="RH_Avg", y="O3", data=df, markers=".", scatter_kws=dict(color='c', s=10))
g.set_xlabels("RH_Avg\n%", fontsize=10)
g.set_ylabels("O3\nppb", fontsize=10)
plt.ylim(0,63)
plt.xlim(0,100)
plt.title("Regression plot between O3 and RH_Avg", fontsize=14, pad=20)
plt.show()
```

## O3 vs Rad_Avg

```python
plt.figure(figsize=(14,12))
g = sns.lmplot(x="Rad_Avg", y="O3", data=df, markers=".", scatter_kws=dict(color='c', s=10))
g.set_xlabels("Rad_Avg\nW/m^2", fontsize=10)
g.set_ylabels("O3\nppb", fontsize=10)
plt.ylim(0,63)
plt.xlim(0,700)
plt.title("Regression plot between O3 and Rad_Avg", fontsize=14, pad=20)
plt.show()
```

## NO2 vs NO

```python
plt.figure(figsize=(14, 12))
g = sns.lmplot(x="NO", y="NO2", data=df, markers=".", scatter_kws=dict(color='chartreuse', s=10), line_kws=dict(color='darkgreen'))
g.set_xlabels("NO\nppb", fontsize=10)
g.set_ylabels("NO2\nppb", fontsize=10)
plt.ylim(0, 17)
plt.xlim(0, 6.5)
plt.title("Regression plot between NO2 and NO", fontsize=14, pad=20)
plt.show()
```

## NO2 and NO vs Wind variables

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(14, 12))

sns.regplot(x="WSpeed_Avg", y="NO2", data=df, ax=axes[0, 0], scatter_kws=dict(color='chartreuse', s=10), line_kws=dict(color='darkgreen'))
axes[0, 0].set_xlabel("WSpeed_Avg\nm/s", fontsize=10)
axes[0, 0].set_ylabel("NO2\nppb", fontsize=10)

sns.regplot(x="WDir_Avg", y="NO2", data=df, ax=axes[0, 1], scatter_kws=dict(color='chartreuse', s=10), line_kws=dict(color='darkgreen'))
axes[0, 1].set_ylabel("NO2\nppb", fontsize=10)
axes[0, 1].set_xlabel("WDir_Avg\ndeg", fontsize=10)

sns.regplot(x="WSpeed_Avg", y="NO", data=df, ax=axes[1, 0], scatter_kws=dict(color='chartreuse', s=10), line_kws=dict(color='darkgreen'))
axes[1, 0].set_ylabel("NO\nppb", fontsize=10)
axes[1, 0].set_xlabel("WSpeed_Avg", fontsize=10)

sns.regplot(x="WDir_Avg", y="NO", data=df, ax=axes[1, 1], scatter_kws=dict(color='chartreuse', s=10), line_kws=dict(color='darkgreen'))
axes[1, 1].set_ylabel("NO\nppb", fontsize=10)
axes[1, 1].set_xlabel("WDir_Avg\ndeg", fontsize=10)

plt.tight_layout()
plt.suptitle("NO2 and NO vs Wind Variables", fontsize=18, y=1.02)
plt.show()
```

## PM10 vs PM2.5

## PM10 vs Temp_Avg

```python
plt.figure(figsize=(14, 12))
g = sns.lmplot(x="Temp_Avg", y="PM10", data=df, markers=".", scatter_kws=dict(color='lightcoral', s=10), line_kws=dict(color='maroon'))
g.set_xlabels("Temp_Avg\n°C", fontsize=10)
g.set_ylabels("PM10\nug/m^3", fontsize=10)
plt.ylim(0, 60)
plt.xlim(0, 30)
plt.title("Regression plot between PM10 and Temp_avg", fontsize=14, pad=20)
plt.show()
```

## PM10 vs RH_Avg

**Strong negative correlation with 99% of confidence**

```python
plt.figure(figsize=(14, 12))
g = sns.lmplot(x="RH_Avg", y="PM10", data=df, markers=".", scatter_kws=dict(color='lightcoral', s=10), line_kws=dict(color='maroon'))
g.set_xlabels("RH_Avg\n%", fontsize=10)
g.set_ylabels("PM10\nug/m^3", fontsize=10)
plt.ylim(0, 60)
plt.xlim(0, 100)
plt.title("Regression plot between PM10 and RH_avg", fontsize=14, pad=20)
plt.show()
```

## PM10 and PM2.5 vs Temp_Avg and RH_Avg

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(14, 12))

sns.regplot(x="Temp_Avg", y="PM10", data=df, ax=axes[0, 0], scatter_kws=dict(color='lightcoral', s=10), line_kws=dict(color='maroon'))
axes[0, 0].set_xlabel("Temp_Avg\n°C", fontsize=10)
axes[0, 0].set_ylabel("PM10\nug/m^3", fontsize=10)

sns.regplot(x="RH_Avg", y="PM10", data=df, ax=axes[0, 1], scatter_kws=dict(color='lightcoral', s=10), line_kws=dict(color='maroon'))
axes[0, 1].set_ylabel("PM10\nug/m^3", fontsize=10)
axes[0, 1].set_xlabel("RH_Avg\n%", fontsize=10)

sns.regplot(x="Temp_Avg", y="PM2.5", data=df, ax=axes[1, 0], scatter_kws=dict(color='lightcoral', s=10), line_kws=dict(color='maroon'))
axes[1, 0].set_ylabel("PM2.5\nug/m^3", fontsize=10)
axes[1, 0].set_xlabel("Temp_Avg\n°C", fontsize=10)

sns.regplot(x="RH_Avg", y="PM2.5", data=df, ax=axes[1, 1], scatter_kws=dict(color='lightcoral', s=10), line_kws=dict(color='maroon'))
axes[1, 1].set_ylabel("PM2.5\nug/m^3", fontsize=10)
axes[1, 1].set_xlabel("RH_Avg\n%", fontsize=10)

plt.tight_layout()
plt.suptitle("PM10 and PM2.5 vs Temp_Avg and RH_Avg", fontsize=18, y=1.02)
plt.show()
```
