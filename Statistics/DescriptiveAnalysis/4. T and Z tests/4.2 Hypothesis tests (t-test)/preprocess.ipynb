Let's create the samples that we need for the hypothesis about O3 concentration during the solar radiation.

- H0: mean(O3_high_radiation) <= mean(O3_low_radiation)
- H1: mean(O3_high_radiation) > mean(O3_low_radiation)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
PATH = "../../../DataCleaning/"
```

```python
df = pd.read_csv(PATH + "Atmospheric Data With No Missing Values.csv")
```

```python
df_high_radiation = df[(df["Hour"] >= 11) & (df["Hour"] <= 16)]
df_low_radiation = df[(df["Hour"] < 11) | (df["Hour"] > 16)]
```

```python
df_high_radiation.to_csv("high_radiation.csv", index=False)
df_low_radiation.to_csv("low_radiation.csv", index=False)
```
