### 3 samples with 10 instances from each day of the year

```python
def everyday_samples(n:int):
    df_subset1, df_subset2, df_subset3 = pd.DataFrame(), pd.DataFrame(), pd.DataFrame()

    for m in range(1, 13):
        if m == 2: day = 30
        elif m == 4 or m == 6 or m == 9 or m == 11: day = 31
        else: day = 32
        for d in range(1, day):
            df_subset = df[(df["Month"] == m) & (df["Day"] == d)]
            if df_subset.shape[0] == 0: continue
            try:
                df_subset1 = pd.concat([df_subset1, df_subset.sample(n=n)], axis=0)
                df_subset2 = pd.concat([df_subset2, df_subset.sample(n=n)], axis=0)
                df_subset3 = pd.concat([df_subset3, df_subset.sample(n=n)], axis=0)
            except:
                df_subset1 = pd.concat([df_subset1, df_subset.sample(n=n, replace=True)], axis=0)
                df_subset2 = pd.concat([df_subset2, df_subset.sample(n=n, replace=True)], axis=0)
                df_subset3 = pd.concat([df_subset3, df_subset.sample(n=n, replace=True)], axis=0)

    return df_subset1, df_subset2, df_subset3
```

```python
df_subset1, df_subset2, df_subset3 = everyday_samples(10)
```

```python
df_subset1
```

*Note:* some days have no instances

```python
df_subset1.to_csv("EverydaySamples10_1.csv", index=False)
df_subset2.to_csv("EverydaySamples10_2.csv", index=False)
df_subset3.to_csv("EverydaySamples10_3.csv", index=False)
```
