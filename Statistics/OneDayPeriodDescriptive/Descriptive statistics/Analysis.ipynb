```python
def fisher(variable: pd.Series) -> float:
    numerator = sum((variable - variable.mean())**3)
    denominator = variable.shape[0] * (variable.std()**3)    
    return numerator / denominator
```

```python
variables = ['O3', 'NO2', 'NO', 'CO', 'PM10', 'PM2.5', 'Temp_Avg', 'RH_Avg', 'WSpeed_Avg', 'WDir_Avg', 'Press_Avg', 'Rad_Avg', 'Rain_Tot']
```

```python
print("Skewness")
for c in variables: print(f"{c:<15} {round(fisher(df[c]), 2):<10}")
```
