```python
import pandas as pd
```

```python
df = pd.read_csv("../Atmospheric Data Cleaned Instances Per Day.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df_subset = df[["Month", "Day", "O3"]]
```

```python
df_subset.head()
```

```python
df_spring_summer = df_subset[((df_subset["Month"] == 3) & (df_subset["Day"] >= 21)) | ((df_subset["Month"] > 3) & (df_subset["Month"] < 9)) | ((df_subset["Month"] == 9) & (df_subset["Day"] < 23))]
df_autumn_winter = df_subset[((df_subset["Month"] < 3) | (df_subset["Month"] > 9)) | ((df_subset["Month"] == 3) & (df_subset["Day"] < 21)) | ((df_subset["Month"] == 9) & (df_subset["Day"] >= 23))]
```

```python
df_spring_summer.sort_values(by=["Month", "Day"])
```

```python
df_spring_summer.to_csv("Ozone_Spring_Summer.csv", index=False)
df_autumn_winter.to_csv("Ozone_Autumn_Winter.csv", index=False)
```
