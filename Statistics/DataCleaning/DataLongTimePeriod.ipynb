```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df_original = pd.read_csv("Atmospheric Data Cleaned - Time Format.csv")
```

```python
df_original.head()
```

```python
df_original.columns
```

```python
df_original.info()
```

```python
df_per_day_preprocess = df_original.groupby(["Year", "Month", "Day"])
```

```python
df_per_day_rain = df_per_day_preprocess["Rain_Tot"].sum()
```

```python
df_per_day_rain.head()
```

```python
df_per_day_other_variables = df_per_day_preprocess[["O3", "NO2", "NO", "CO", "PM10", "PM2.5", "Temp_Avg", "RH_Avg", "WSpeed_Avg", "WDir_Avg", "Press_Avg", "Rad_Avg"]].mean()
```

```python
df_per_day_other_variables.head()
```

```python
df_per_day = pd.merge(df_per_day_other_variables, df_per_day_rain, on=["Year", "Month", "Day"], how='inner').drop_duplicates()
```

```python
df_per_day.head()
```

```python
df_per_day.info()
```

```python
df_per_day["Rad_Avg"] = df_per_day["Rad_Avg"].fillna(df_per_day["Rad_Avg"].median())
```

```python
df_per_day.info()
```

```python
df_per_day.head()
```

```python
df_per_day.to_csv("Atmospheric Data Cleaned Instances Per Day.csv")
```
