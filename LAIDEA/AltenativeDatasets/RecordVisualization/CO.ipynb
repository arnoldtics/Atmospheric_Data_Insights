```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
PATH = "../Datasets/"
pd.set_option("display.max_columns", None)
sns.set_style("darkgrid")
```

```python
df = pd.read_csv(PATH + "CompleteDataset.csv")
df["Fecha"] = pd.to_datetime(df["Fecha"])
```

```python
df["CO"].describe()
```

```python
def clean(x):
    if x >= 0.04 and x <= 25: return x
    return np.nan
```

```python
df["CO"] = df["CO"].apply(clean)
```

```python
df["CO"].info()
```

```python
plt.figure(figsize=(16, 8))
sns.scatterplot(x=df["Fecha"], y=df["CO"], s=10, color="olive")
# plt.ylim(0, 140)
plt.ylabel("CO concentration\nppm", size=16)
plt.xlabel("Time", size=16)
plt.title("Morelia's CO concentration per day from 2018 to 2025\nDowntown dataset", size=20)
plt.show()
```

```python
plt.figure(figsize=(16, 8))
sns.scatterplot(x=df["Fecha"], y=df["CO"], s=10, color="olive")
plt.ylim(0, 3)
plt.ylabel("CO concentration\nppm", size=16)
plt.xlabel("Time", size=16)
plt.title("Morelia's CO concentration per day from 2018 to 2025\nDowntown dataset", size=20)
plt.show()
```
