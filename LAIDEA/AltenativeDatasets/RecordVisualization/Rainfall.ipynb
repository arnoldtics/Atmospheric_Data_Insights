```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
PATH = "../Datasets/"
pd.set_option("display.max_columns", None)
sns.set_style("darkgrid")
```

```python
df = pd.read_csv(PATH + "CompleteDataset.csv")
df["Fecha"] = pd.to_datetime(df["Fecha"])
```

```python
df["Rainfall"].describe()
```

```python
plt.figure(figsize=(16, 8))
sns.scatterplot(x=df["Fecha"], y=df["Rainfall"], s=10, color="dodgerblue")
# plt.ylim(-1, 55)
plt.ylabel("Precipitation\nmm", size=16)
plt.xlabel("Time", size=16)
plt.title("Morelia's rainfall per day from 2018 to 2025\nDowntown dataset", size=20)
plt.show()
```

```python
plt.figure(figsize=(16, 8))
sns.scatterplot(x=df["Fecha"], y=df["Rainfall"], s=10, color="dodgerblue")
plt.ylim(-1, 20)
plt.ylabel("Precipitation\nmm", size=16)
plt.xlabel("Time", size=16)
plt.title("Morelia's rainfall per day from 2018 to 2025\nDowntown dataset", size=20)
plt.show()
```
