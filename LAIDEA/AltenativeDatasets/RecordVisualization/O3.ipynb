```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
PATH = "../Datasets/"
pd.set_option("display.max_columns", None)
sns.set_style("darkgrid")
```

```python
df = pd.read_csv(PATH + "CompleteDataset.csv")
df["Fecha"] = pd.to_datetime(df["Fecha"])
```

```python
df["O3"].describe()
```

```python
def clean(x):
    if x >= 0.03 and x <= 500: return x
    return np.nan
```

```python
df["O3"] = df["O3"].apply(clean)
```

```python
df["O3"].info()
```

```python
plt.figure(figsize=(16, 8))
sns.scatterplot(x=df["Fecha"], y=df["O3"], s=10, color="royalblue")
plt.ylim(0, 140)
plt.ylabel("O3 concentration\nppb", size=16)
plt.xlabel("Time", size=16)
plt.title("Morelia's O3 concentration per day from 2018 to 2025\nDowntown dataset", size=20)
plt.show()
```
