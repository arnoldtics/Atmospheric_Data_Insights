```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
PATH = "../Datasets/"
pd.set_option("display.max_columns", None)
sns.set_style("darkgrid")
```

```python
df = pd.read_csv(PATH + "CompleteDataset.csv")
df["Fecha"] = pd.to_datetime(df["Fecha"])
```

```python
df["NO2"].describe()
```

```python
def clean(x):
    if x >= 0.04 and x <= 3000: return x
    return np.nan
```

```python
df["NO2"] = df["NO2"].apply(clean)
```

```python
df["NO2"].info()
```

```python
plt.figure(figsize=(16, 8))
sns.scatterplot(x=df["Fecha"], y=df["NO2"], s=10, color="limegreen")
plt.ylim(0, 390)
plt.ylabel("NO2 concentration\nppb", size=16)
plt.xlabel("Time", size=16)
plt.title("Morelia's NO2 concentration per day from 2018 to 2025\nDowntown dataset", size=20)
plt.show()
```
