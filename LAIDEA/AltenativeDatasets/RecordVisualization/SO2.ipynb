```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
PATH = "../Datasets/"
pd.set_option("display.max_columns", None)
sns.set_style("darkgrid")
```

```python
df = pd.read_csv(PATH + "CompleteDataset.csv")
df["Fecha"] = pd.to_datetime(df["Fecha"])
```

```python
df["SO2"].describe()
```

```python
def clean(x):
    if x >= 0.5 and x <= 1000: return x
    return np.nan
```

```python
df["SO2"] = df["SO2"].apply(clean)
```

```python
df["SO2"].info()
```

```python
plt.figure(figsize=(16, 8))
sns.scatterplot(x=df["Fecha"], y=df["SO2"], s=10, color="peru")
plt.ylim(0, 75)
plt.ylabel("SO2 concentration\nppb", size=16)
plt.xlabel("Time", size=16)
plt.title("Morelia's SO2 concentration per day from 2018 to 2020\nDowntown dataset", size=20)
plt.show()
```
