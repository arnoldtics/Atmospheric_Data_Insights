```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
PATH = "../Datasets/"
pd.set_option("display.max_columns", None)
sns.set_style("darkgrid")
```

```python
df = pd.read_csv(PATH + "CompleteDataset.csv")
df["Fecha"] = pd.to_datetime(df["Fecha"])
```

```python
df["PM10"].describe()
```

```python
def clean(x):
    if x >= 4 and x <= 400: return x
    return np.nan
```

```python
df["PM10"] = df["PM10"].apply(clean)
```

```python
df["PM10"].info()
```

```python
plt.figure(figsize=(16, 8))
sns.scatterplot(x=df["Fecha"], y=df["PM10"], s=10, color="crimson")
# plt.ylim(0, 34)
plt.ylabel("PM10 concentration\ng/m", size=16)
plt.xlabel("Time", size=16)
plt.title("Morelia's PM10 concentration per day from 2018 to 2025\nDowntown dataset", size=20)
plt.show()
```
