```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
PATH = "../Datasets/"
pd.set_option("display.max_columns", None)
sns.set_style("darkgrid")
```

```python
df = pd.read_csv(PATH + "CompleteDataset.csv")
df["Fecha"] = pd.to_datetime(df["Fecha"])
```

```python
def clean(x):
    if x >= 4: return x
    return np.nan
```

```python
df["PM2.5"] = df["PM2.5"].apply(clean)
```

```python
df["PM2.5"].describe()
```

```python
df["WindDirection"].describe()
```

```python
df["WindDirectionRad"] = np.radians(df["WindDirection"])
```

```python
fig, ax = plt.subplots(subplot_kw={'projection':'polar'})
ax.scatter(df["WindDirectionRad"], df["PM2.5"], s=0.5, color='chocolate')

ax.set_theta_zero_location('N')  # 0 en la parte superior
ax.set_theta_direction(-1)  # Sentido horario

ax.set_title("Relationship between wind direction and PM2.5 concentration in Morelia\nDowntown dataset", pad=20)
ax.set_xlabel("Wind Direction ()") 
ax.set_ylabel("PM2.5 Concentration (g/m)", labelpad=40)
ax.set_xticks(np.radians([0, 45, 90, 135, 180, 225, 270, 315]))
ax.set_xticklabels(['N', 'NE', 'E', 'SE', 'S', 'SW', 'W', 'NW'])

plt.show()
```
