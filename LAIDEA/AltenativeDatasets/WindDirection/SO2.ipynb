```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
PATH = "../Datasets/"
pd.set_option("display.max_columns", None)
sns.set_style("darkgrid")
```

```python
df = pd.read_csv(PATH + "CompleteDataset.csv")
df["Fecha"] = pd.to_datetime(df["Fecha"])
```

```python
def clean(x):
    if x >= 0.5 and x <= 1000: return x
    return np.nan
```

```python
df["SO2"] = df["SO2"].apply(clean)
```

```python
df["SO2"].describe()
```

```python
df["WindDirection"].describe()
```

```python
df["WindDirectionRad"] = np.radians(df["WindDirection"])
```

```python
fig, ax = plt.subplots(subplot_kw={'projection':'polar'})
ax.scatter(df["WindDirectionRad"], df["SO2"], s=0.5, color='peru')

ax.set_theta_zero_location('N')  # 0° en la parte superior
ax.set_theta_direction(-1)  # Sentido horario

ax.set_title("Relationship between wind direction and SO2 concentration in Morelia\nDowntown dataset", pad=20)
ax.set_xlabel("Wind Direction (°)") 
ax.set_ylabel("SO2 Concentration (ppb)", labelpad=40)
ax.set_xticks(np.radians([0, 45, 90, 135, 180, 225, 270, 315]))
ax.set_xticklabels(['N', 'NE', 'E', 'SE', 'S', 'SW', 'W', 'NW'])

plt.show()
```

```python
fig, ax = plt.subplots(subplot_kw={'projection':'polar'})
ax.scatter(df["WindDirectionRad"], df["NO2"], s=0.5, color='forestgreen')

ax.set_theta_zero_location('N')  # 0° en la parte superior
ax.set_theta_direction(-1)  # Sentido horario

ax.set_title("Relationship between wind direction and NO2 concentration in Morelia\nDowntown dataset", pad=20)
ax.set_xlabel("Wind Direction (°)") 
ax.set_ylabel("NO2 Concentration (ppb)", labelpad=40)
ax.set_xticks(np.radians([0, 45, 90, 135, 180, 225, 270, 315]))
ax.set_xticklabels(['N', 'NE', 'E', 'SE', 'S', 'SW', 'W', 'NW'])
ax.set_ylim(0, 60)

plt.show()
```
