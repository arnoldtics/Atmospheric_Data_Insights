```python
import pandas as pd
import os
from functools import reduce
```

```python
PATH = "../Datasets/"
pd.set_option("display.max_columns", None)
```

```python
variables = ["Temperature", "RelativeHumidity", "SolarRadiation", "Rainfall", "WindDirection",
             "PM10", "PM2.5", "O3", "CO", "NO", "NO2", "SO2"]
```

```python
df = pd.read_csv(PATH + variables[0] + "/" + variables[0] + ".csv")
df["Fecha"] = pd.to_datetime(df["Fecha"])
```

```python
df
```

```python
dfs = []
for title in variables:
    df = pd.read_csv(PATH + title + "/" + title + ".csv")
    df["Fecha"] = pd.to_datetime(df["Fecha"])
    dfs.append(df)
```

```python
df = reduce(lambda left, right: pd.merge(left, right, on="Fecha", how="outer"), dfs)
```

```python
df
```

```python
df.info()
```

```python
full_range = pd.date_range(start=df["Fecha"].min(), end=df["Fecha"].max(), freq='D')
dfFull = pd.DataFrame(full_range, columns=["Fecha"])
dfFinal = dfFull.merge(df, on="Fecha", how="left")
```
